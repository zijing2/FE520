```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
plt.plot(walk)

# print(np.where(abs(walk)>10, 10, abs(walk)))

n = np.argmax(np.where(abs(walk)>30, 30, abs(walk)))
print (n);
```
